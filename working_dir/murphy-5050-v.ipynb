## Neural Network 2: 6-Layer with 8, 16, 21, 32 Neurons (UNCLEANED)

---

```python
# Imports
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
# Load the cleaned 50/50 dataset
df = pd.read_csv('diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
```

```python
# Split into features & target, then train/test (50:50)
X = df.drop('Diabetes_binary', axis=1).values
y = df['Diabetes_binary'].values

# Single train/test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Scale features
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)

# Model builder for 6-layer net (4 hidden + in/out)
def build_model(units):
    m = Sequential([
        Dense(units, activation='relu', input_dim=X_train.shape[1]),
        Dense(units, activation='relu'),
        Dense(units, activation='relu'),
        Dense(units, activation='relu'),
        Dense(1,     activation='sigmoid'),
    ])
    m.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    return m


# Train & evaluate each variant, no extra splits
results = {}
for units in [8, 16, 21, 32]:
    print(f"\nTraining with {units} neurons per hidden layer")
    model = build_model(units)
    model.fit(X_train, y_train,
              epochs=30,
              batch_size=32,
              verbose=1)      # no validation_split here
    
    preds = (model.predict(X_test) > 0.5).astype(int).flatten()
    acc = accuracy_score(y_test, preds)
    print(f"Test accuracy: {acc:.4f}")
    results[units] = acc

# Cell 6 – Comparison plot
plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('Neurons per hidden layer')
plt.ylabel('Test Accuracy')
plt.title('Model Size vs. Accuracy')
plt.xticks([8,16,21,32])
plt.show()
```

## 6x16 Network

---

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, ConfusionMatrixDisplay

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.callbacks import EarlyStopping

import matplotlib.pyplot as plt

# load cleaned 50/50 dataset
df = pd.read_csv("5050_clean_diabetes_dataset.csv")
X = df.drop("Diabetes_binary", axis=1).values
y = df["Diabetes_binary"].values
```

```python
# 80/20 split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# scale features (fit on train only)
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
def make_model(input_dim, dropout_rate=0.3):
    m = Sequential()
    for i in range(6):
        if i == 0:
            m.add(Dense(16, input_dim=input_dim, use_bias=False))
        else:
            m.add(Dense(16, use_bias=False))
        m.add(BatchNormalization())
        m.add(tf.keras.layers.ReLU())
        m.add(Dropout(dropout_rate))
    m.add(Dense(1, activation="sigmoid"))
    m.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return m

# instantiate
model = make_model(input_dim=X_train.shape[1], dropout_rate=0.3)
model.summary()
```

```python
es = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=100,
    batch_size=32,
    callbacks=[es],
    verbose=2
)
```

```python
y_prob = model.predict(X_test).flatten()
y_pred = (y_prob >= 0.5).astype(int)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize="true")
plt.show()

# ROC & AUC
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="val loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()
plt.title("Training vs. Validation Loss"); plt.show()

plt.plot(history.history["accuracy"], label="train acc")
plt.plot(history.history["val_accuracy"], label="val acc")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
plt.title("Training vs. Validation Accuracy"); plt.show()
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    roc_curve, auc, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# load cleaned dataset
df = pd.read_csv("5050_clean_diabetes_dataset.csv")
X = df.drop("Diabetes_binary", axis=1).values.astype(np.float32)
y = df["Diabetes_binary"].values.astype(int)
```

```python
# Cell 2 – Train/Test Split & Scaling
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
# Cell 3 – Define base MLPClassifier
from sklearn.neural_network import MLPClassifier

base_mlp = MLPClassifier(
    hidden_layer_sizes=(16,)*6,   # six layers of 16 units
    activation="relu",
    solver="adam",
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=10,
    random_state=42,
    max_iter=200
)
```

```python
param_grid = {
    "hidden_layer_sizes": [(8,)*6, (16,)*6, (32,)*6],
    "alpha":              [1e-4, 1e-3, 1e-2],      # L2 penalty
    "learning_rate_init": [1e-3, 1e-2],
    "batch_size":         [32, 64],
}

grid = GridSearchCV(
    estimator=base_mlp,
    param_grid=param_grid,
    cv=3,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2
)
grid.fit(X_train, y_train)
```

```python
print("Best params:      ", grid.best_params_)
print("Best CV accuracy: ", grid.best_score_)
```

```python
best_mlp = grid.best_estimator_
y_prob  = best_mlp.predict_proba(X_test)[:, 1]
y_pred  = best_mlp.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curve"); plt.legend(); plt.show()

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize="true")
plt.show()
```

## Grid search with activaiton functions

```python
# Re-importing necessary libraries after environment reset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Load the uncleaned 50/50 diabetes dataset
df = pd.read_csv('5050_clean_diabetes_dataset.csv')
X = df.drop('Diabetes_binary', axis=1).values
y = df['Diabetes_binary'].values

# Single 80/20 train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Standardize features
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)

# Define hyperparameter grid
learning_rates = [5e-3, 1e-2, 2e-2]
alphas = [0.0, 1e-4]
batch_sizes = [8, 16, 32]
activations = ['relu', 'tanh', 'logistic']  # No LeakyReLU directly in MLPClassifier

results = []

# Manual 3-fold CV to avoid timeout issues
kf = KFold(n_splits=3, shuffle=True, random_state=42)

for lr in learning_rates:
    for alpha in alphas:
        for bs in batch_sizes:
            for activation in activations:
                fold_accuracies = []
                for train_index, val_index in kf.split(X_train):
                    X_tr, X_val = X_train[train_index], X_train[val_index]
                    y_tr, y_val = y_train[train_index], y_train[val_index]

                    # Build the model
                    mlp = MLPClassifier(
                        hidden_layer_sizes=(16, 16, 16, 16),
                        activation=activation,
                        solver='adam',
                        learning_rate_init=lr,
                        alpha=alpha,
                        batch_size=bs,
                        max_iter=200,
                        random_state=42
                    )
                    # Train the model
                    mlp.fit(X_tr, y_tr)

                    # Validate
                    val_acc = mlp.score(X_val, y_val)
                    fold_accuracies.append(val_acc)

                # Record the mean accuracy across the 3 folds
                avg_acc = np.mean(fold_accuracies)
                results.append({
                    'learning_rate': lr,
                    'alpha': alpha,
                    'batch_size': bs,
                    'activation': activation,
                    'mean_cv_acc': avg_acc
                })

# Convert results to a DataFrame for easier viewing
results_df = pd.DataFrame(results)
best_params = results_df.loc[results_df['mean_cv_acc'].idxmax()]

# Test on the held-out test set with the best parameters
best_mlp = MLPClassifier(
    hidden_layer_sizes=(16, 16, 16, 16),
    activation=best_params['activation'],
    solver='adam',
    learning_rate_init=best_params['learning_rate'],
    alpha=best_params['alpha'],
    batch_size=int(best_params['batch_size']),
    max_iter=200,
    random_state=42
)
best_mlp.fit(X_train, y_train)
test_acc = best_mlp.score(X_test, y_test)

best_params, test_acc
```

```python
# Now regenerate the visualizations based on the freshly computed results_df

# --- 1. Bar Plot: Mean Accuracy per Activation Function ---
plt.figure(figsize=(8, 5))
sns.barplot(data=results_df, x='activation', y='mean_cv_acc', ci=None)
plt.title('Mean CV Accuracy by Activation Function')
plt.ylabel('Mean CV Accuracy')
plt.xlabel('Activation Function')
plt.ylim(0.7, 0.78)
plt.grid(True)
plt.tight_layout()
plt.show()

# --- 2. Heatmap: Accuracy by Learning Rate vs Batch Size (ReLU Only) ---
relu_df = results_df[results_df['activation'] == 'relu']
heatmap_data = relu_df.pivot_table(index='batch_size', columns='learning_rate', values='mean_cv_acc')

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".3f", cmap='viridis')
plt.title('CV Accuracy Heatmap (ReLU): Learning Rate vs Batch Size')
plt.xlabel('Learning Rate')
plt.ylabel('Batch Size')
plt.tight_layout()
plt.show()

```
