## Neural Network 2: 6-Layer with 8, 16, 21, 32 Neurons (UNCLEANED)

---

```python
# Cell 1 – Imports
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
# Cell 2 – Load the cleaned 50/50 dataset
df = pd.read_csv('diabetes_binary_5050split_health_indicators_BRFSS2015.csv')
```

```python
# Cell 3 – Split into features & target, then train/test (50:50)
X = df.drop('Diabetes_binary', axis=1).values
y = df['Diabetes_binary'].values

# Cell 3 – Single train/test split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

# Scale features
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)

# Cell 4 – Model builder for 6-layer net (4 hidden + in/out)
def build_model(units):
    m = Sequential([
        Dense(units, activation='relu', input_dim=X_train.shape[1]),
        Dense(units, activation='relu'),
        Dense(units, activation='relu'),
        Dense(units, activation='relu'),
        Dense(1,     activation='sigmoid'),
    ])
    m.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
    return m


# Cell 5 – Train & evaluate each variant, no extra splits
results = {}
for units in [8, 16, 21, 32]:
    print(f"\nTraining with {units} neurons per hidden layer")
    model = build_model(units)
    model.fit(X_train, y_train,
              epochs=30,
              batch_size=32,
              verbose=1)      # no validation_split here
    
    preds = (model.predict(X_test) > 0.5).astype(int).flatten()
    acc = accuracy_score(y_test, preds)
    print(f"Test accuracy: {acc:.4f}")
    results[units] = acc

# Cell 6 – Comparison plot
plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('Neurons per hidden layer')
plt.ylabel('Test Accuracy')
plt.title('Model Size vs. Accuracy')
plt.xticks([8,16,21,32])
plt.show()
```

## 6x16 Network

---

```python
# Cell 1
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, ConfusionMatrixDisplay

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras.callbacks import EarlyStopping

import matplotlib.pyplot as plt

# load cleaned 50/50 dataset
df = pd.read_csv("5050_clean_diabetes_dataset.csv")
X = df.drop("Diabetes_binary", axis=1).values
y = df["Diabetes_binary"].values
```

```python
# Cell 2
# 80/20 split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# scale features (fit on train only)
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
# Cell 3
def make_model(input_dim, dropout_rate=0.3):
    m = Sequential()
    for i in range(6):
        if i == 0:
            m.add(Dense(16, input_dim=input_dim, use_bias=False))
        else:
            m.add(Dense(16, use_bias=False))
        m.add(BatchNormalization())
        m.add(tf.keras.layers.ReLU())
        m.add(Dropout(dropout_rate))
    m.add(Dense(1, activation="sigmoid"))
    m.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )
    return m

# instantiate
model = make_model(input_dim=X_train.shape[1], dropout_rate=0.3)
model.summary()
```

```python
# Cell 4
es = EarlyStopping(
    monitor="val_loss",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train, y_train,
    validation_split=0.1,
    epochs=100,
    batch_size=32,
    callbacks=[es],
    verbose=2
)
```

```python
# Cell 5
# raw predictions
y_prob = model.predict(X_test).flatten()
y_pred = (y_prob >= 0.5).astype(int)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize="true")
plt.show()

# ROC & AUC
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Cell 6
plt.plot(history.history["loss"], label="train loss")
plt.plot(history.history["val_loss"], label="val loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.legend()
plt.title("Training vs. Validation Loss"); plt.show()

plt.plot(history.history["accuracy"], label="train acc")
plt.plot(history.history["val_accuracy"], label="val acc")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.legend()
plt.title("Training vs. Validation Accuracy"); plt.show()
```

## With updated hyperparameters

```python
# Cell 1 – Imports & Data Load
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, classification_report,
    roc_curve, auc, ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

# load cleaned dataset
df = pd.read_csv("5050_clean_diabetes_dataset.csv")
X = df.drop("Diabetes_binary", axis=1).values.astype(np.float32)
y = df["Diabetes_binary"].values.astype(int)
```

```python
# Cell 2 – Train/Test Split & Scaling
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
# Cell 3 – Define base MLPClassifier
from sklearn.neural_network import MLPClassifier

base_mlp = MLPClassifier(
    hidden_layer_sizes=(16,)*6,   # six layers of 16 units
    activation="relu",
    solver="adam",
    early_stopping=True,
    validation_fraction=0.1,
    n_iter_no_change=10,
    random_state=42,
    max_iter=200
)
```

```python
# Cell 4 – GridSearchCV over key hyperparameters
param_grid = {
    "hidden_layer_sizes": [(8,)*6, (16,)*6, (32,)*6],
    "alpha":              [1e-4, 1e-3, 1e-2],      # L2 penalty
    "learning_rate_init": [1e-3, 1e-2],
    "batch_size":         [32, 64],
}

grid = GridSearchCV(
    estimator=base_mlp,
    param_grid=param_grid,
    cv=3,
    scoring="accuracy",
    n_jobs=-1,
    verbose=2
)
grid.fit(X_train, y_train)
```

```python
# Cell 5 – Best Hyperparameters & CV Score
print("Best params:      ", grid.best_params_)
print("Best CV accuracy: ", grid.best_score_)
```

```python
# Cell 6 – Final Test‐Set Evaluation & Plots
best_mlp = grid.best_estimator_
y_prob  = best_mlp.predict_proba(X_test)[:, 1]
y_pred  = best_mlp.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1],[0,1],"--")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.title("ROC Curve"); plt.legend(); plt.show()

# Confusion Matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize="true")
plt.show()
```
